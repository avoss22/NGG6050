**This code simulates the results from Bollmann et al. Nature Neuroscience 2023  Figure 1f, 1 and 2 **

"To better understand the mesoscopic functional organization of these developing circuits, we next analyzed the distributions of both functional output and input connectivity degrees across developmental stages (Fig. 1f, 1 and 2, respectively). To test the overall distribution that best fits the data, we followed a statistical procedure outlined in ref. 23 (Methods). In this way, we could determine whether input and output functional connectivity degrees followed a power law, exponential or log-normal distribution5,24,25,26,27. We found that both the input and output connectivity degrees were best fitted with a log-normal distribution with the following parameters (output connectivity: mean = 1.01, s.d. = 1.18; input connectivity: mean = 1.04, s.d. = 1.21)."

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the log-normal distribution for output connectivity
output_mean = 1.01
output_stddev = 1.18

# Parameters of the log-normal distribution for input connectivity
input_mean = 1.04
input_stddev = 1.21

# Set seed for reproducibility
np.random.seed(42)

# Simulate the log-normal distribution for output connectivity
output_samples = np.random.lognormal(mean=np.log(output_mean), sigma=output_stddev, size=10000)

# Simulate the log-normal distribution for input connectivity
input_samples = np.random.lognormal(mean=np.log(input_mean), sigma=input_stddev, size=10000)

# Plot histograms to visualize the distributions
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1) #Creates 2 plots in 1 figure (this figure has 1 row, 2 columns, and this is the first plot)
plt.hist(output_samples, bins=50, density=True, alpha=0.6, color='b', label='Output Connectivity')
plt.title('Output Connectivity Distribution')
plt.xlabel('Output Connectivity (% active)')
plt.ylabel('Probability %')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(input_samples, bins=50, density=True, alpha=0.6, color='g', label='Input Connectivity')
plt.title('Input Connectivity Distribution')
plt.xlabel('Input Connectivity (% active)')
plt.ylabel('Probability %')
plt.legend()

plt.tight_layout()
plt.show()
```
